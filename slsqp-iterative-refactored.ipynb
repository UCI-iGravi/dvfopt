# Iterative SLSQP — Refactored
Uses `modules.dvfopt` for all correction logic. Same test cases as `slsqp-iterative.ipynb`.

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(linewidth=1000)

import modules.laplacian as laplacian
import modules.jacobian as jacobian
from modules.dvfopt import (
    iterative_with_jacobians2,
    plot_deformations,
    run_lapl_and_correction,
    jacobian_det2D,
)
```

# Test cases
## Standard run

```python
msample = np.array([
    [0, 2, 2],
    [0, 3, 5],
])
fsample = np.array([
    [0, 2, 5],
    [0, 3, 2],
])

fixed_sample = np.zeros((1, 10, 10)) # depth, rows, cols

run_lapl_and_correction(fixed_sample, msample, fsample)
```

```python
msample = np.array([
    [0, 0, 0],
    [0, 4, 0],
])
fsample = np.array([
    [0, 3, 3],
    [0, 1, 3],
])

fixed_sample = np.zeros((1, 10, 10))
deformation_i, A, Zd, Yd, Xd = laplacian.sliceToSlice3DLaplacian(fixed_sample, msample, fsample)
print(f'deformation shape: {deformation_i.shape}')
print(deformation_i)
phi_corrected = iterative_with_jacobians2(deformation_i, methodName='SLSQP')
plot_deformations(msample, fsample, deformation_i, phi_corrected, figsize=(10,10))
print(jacobian_det2D(deformation_i[1:]))
```
